Kommentare zur Matrix
----------------------------

- Konflikte entstehen, wenn zwei Indikatoren für zwei Objekte gegenläufig reagieren.
- Bei 4 Indikatoren kann man 4*3/2 Indikatorpaare bilden und herausfinden, 
  welches Paar für ein bestimmtes Paar von Objekten einen Konflikt hervorruft.

Beispiel
--------
Nationen mit der ID 1 und 3: Indikatorpaar q1,q2 verursacht keinen Konflikt. 
Aber in Bezug auf das Indikatorenpaar q1,q4 wird ein Konflikt evident. Daher ist der Eintrag in der ACM gleich 1
Die rowsum summiert die Zahl der Konflikte für ein bestehendes Paar von Objekten. Ist diese Zahl hoch, dann bewirkt der Indikatorsatz q1,…,q4 (Summe über alle Zeilen = 3) besonders viele Konflikte.
Die rowsum ist für das Nationenpaar 3,4 gleich Null, also keine Konflikte und damit vergleichbar. 

Wie das Hassediagramm zeigt, sind 3-4 vergleichbar. Hier gibt es überhaupt keinen Konflikt. 

Man kann auch umgekehrt schauen ob ein betsimmtes Indikatorenpaar besonders viele Konflikte unter den Nationen (Objekten) bewirkt. Beispiel q1,q4:  Das Objakt(Nationen)paar 1,3 hat konfligierende Werte in bezug auf q1, q4
Dagegen wird das Objektpaar 3,2 von diesem speziellen Indikatorpaar nicht in einen Konflikt „gestürzt“. 
Insgesamt zeigt die colsum an, wie gravierend ein bestimmtes Indikatorpaar für Konflikte in der Objektmenge sorgt.  Demnach ist q3,q4 besonders pathologisch. Es verursacht die meisten Konflikte für die Nationenpaare.
